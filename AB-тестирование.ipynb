*После окончания А/В-теста вы получили файл с выгрузкой результатов (файл Data for results.csv). Рассчитайте статистическую значимость различий между группами.
Сделайте выводы по проведенному расчету и интерпретируйте полученный результат.*

Для расчета статистической значимости также можно воспользоваться библиотеками Python. Для анализа подгрузим дополнительную библиотеку statmodels.

```python
from statsmodels.stats.proportion import proportions_ztest, proportion_confint
ab_test = pd.read_csv("Data for results.csv")
```

Далее проверим данные на наличие повторных сессий пользоватлями. И если такие существуют, уберем их из данных. 

```python
session_counts = ab_test['user_id'].value_counts(ascending=False)
multi_users = session_counts[session_counts > 1].count()
print(f'В датасете {multi_users} пользователей, которые встречаются несколько раз')
```

Собственно, повторных сессий у нас в данных нет, можем приступать к работе с этими данными. 

```python
convert = ab_test.groupby('converted')['group'].value_counts().reset_index()
control = convert[convert['group'] == "control"]
treatment = convert[convert['group'] == "treatment"]
conv_control = (control.iloc[1]['count'] / control.iloc[0]['count']*100).round(2)
conv_treatment = (treatment.iloc[1]['count'] / treatment.iloc[0]['count']*100).round(2)
print(f'Конверсия тестовой группы составила {conv_treatment}%')
print(f'Конверсия контрольной группы составила {conv_control}%')
```

По результатам A/B теста видно, что конверсия тестовой группы составила **14,51%**, а конверсия контрольной группы составида **11,05%**. Несмотря на рост конверсии почти на **3,5%**, для более полного анализа результатов теста стоит провести тест Стьюдента, чтобы оценить **значимость различий** двух выборок и поняять, вызваны ли изменения в конверсии качественными изменениями мобильной версии сайта или вызваны внешними эффектами (например, разном количестве пользователей, которые принимали участие в тесте в разных группах).

Далее разобьем данные о пользователях на тестовую и контрольную группы, просуммируем успешные результаты, когда пользователи совершили целевое действие и полученные результаты используем в функции из библиотеки **statsmodels**, которая проводит **t-критерий Стьюдента.** 

```python
control_results = ab_test[ab_test['group'] == 'control']['converted']
treatment_results = ab_test[ab_test['group'] == 'treatment']['converted']
n_con = control_results.count()
n_treat = treatment_results.count()
successes = [control_results.sum(), treatment_results.sum()]
nobs = [n_con, n_treat]
z_stat, pval = proportions_ztest(successes, nobs=nobs)
(lower_con, lower_treat), (upper_con, upper_treat) = proportion_confint(successes, nobs=nobs, alpha=0.05)
print(f'z statistic: {z_stat:.2f}')
print(f'p-value: {pval:.3f}')
print(f'ci 95% for control group: [{lower_con:.3f}, {upper_con:.3f}]')
print(f'ci 95% for treatment group: [{lower_treat:.3f}, {upper_treat:.3f}]')
```

Проведенный тест Стьюдента показал, что p-value гораздо больше порога значения в 0,05. Это говорит нам о том, что мы не можем отвергнуть гипотезу о равенстве выборок. Т.е. **группы статистически значимо не отличаются.**

Таким образом, можно сделать общий вывод по **проведенному тестированию.**

Была сформирована **гипотеза**, которая предлагала улучшение мобильной версии сайта, и имела высокий потенциальный эффект, высокую важность для будущих задач бизнеса и относительно высокий уровень уверенности в правильности гипотезы. 

По результатам А/В-теста по улучшению мобильной версии сайта, было обнаружено, что между контрольной и тестовой группой **отсутствует статистическая значимость**. Это означает, что изменения, внесенные в мобильную версию сайта, не привели к статистически значимому улучшению или изменению в целевых метриках, которые были измерены в рамках теста.
